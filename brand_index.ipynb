```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

mpl.rc('font', family='Malgun Gothic')
```

```python
brand = ['페넬로페', '하기스', '팸퍼스', '리베로', '킨도']
join_index = [194833, 212851, 193356, 143664, 148092]
talk_index = [320425, 249339, 237873, 184471, 186913]
community_index = [200087, 227844, 200097, 189728, 172982]
brand_index = [715345, 690034, 631326, 517863, 507987]
pre_brand_index = [875741, 851033, 510607, 314246, 514919]

df1 = pd.DataFrame([brand, join_index, talk_index, community_index])
df2 = pd.DataFrame([pre_brand_index, brand_index])
df1 = df1.T
df2 = df2.T
```

```python
df1 = df1.rename(columns = {0 : '기저귀 브랜드', 1 : '참여지수', 2 : '소통지수', 3 : '커뮤니티지수'})
df2 = df2.rename(columns = {0 : '5월 브랜드평판지수', 1 : '6월 브랜드평판지수'})
```

```python
df_concat = pd.concat([df1, df2], axis = 1)
df_concat
```

```python
df_concat.plot(x = '기저귀 브랜드', y = ['5월 브랜드평판지수', '6월 브랜드평판지수'], kind = 'bar', color = ['pink', 'm'])
plt.xticks(rotation = 0)
plt.ylim([0, 900000])
plt.show()
```

```python
df_concat.plot(x = '기저귀 브랜드', y = ['6월 브랜드평판지수', '5월 브랜드평판지수'], kind = 'bar', color = ['m', 'pink'])
plt.xticks(rotation = 0)
# plt.ylim([400000, 1400000])
plt.show()
```

```python
df2 = df_concat[(df_concat['기저귀 브랜드'] == '페넬로페') | (df_concat['기저귀 브랜드'] == '하기스') | (df_concat['기저귀 브랜드'] == '팸퍼스')]
df2
```

```python
sns.set_palette("Paired", 6)
sns.barplot(data = df2, x = '기저귀 브랜드', y = '커뮤니티지수')
plt.legend(labels = ['커뮤니티지수'])
```
